### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
# 記得先 Import 正確的套件
import numpy as np
```

```python
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
```

```python
%timeit -n 10 a = 2

for column in Z:
    for row in column:
        for i in row:
            i +=1
Z
```

```python
%timeit -n 10 a = 2

Z+1
```

```python
%timeit -n 10 a = 2

for i in np.array(np.nditer(Z)):
    i += 1
Z
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python

Z = np.random.randint(0,10,50)
print(Z)
```

```python
rint = np.argmax(np.bincount(Z))
rint
```

```python
np.where(Z==rint)
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))


b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```
